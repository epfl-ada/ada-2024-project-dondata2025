```python
import pandas as pd

DATA_FOLDER = 'data/imdb/'

# Reading title basics
title_basics = pd.read_csv(DATA_FOLDER+'title.basics.tsv', sep='\t', low_memory=False)
# Keeping only movies
title_basics = title_basics[title_basics['titleType'] == 'movie']
print(f"There are {len(title_basics)} movies in the dataset.")
title_basics.head()
```

```python
# Reading title ratings
title_ratings = pd.read_csv(DATA_FOLDER+'title.ratings.tsv', sep='\t', low_memory=False)
title_ratings.head()
```

```python
# Merging title basics and title ratings, to keep only primary title, original title, number of votes and average rating
title_basics_ratings = pd.merge(title_basics, title_ratings, on='tconst', how='inner')
title_basics_ratings = title_basics_ratings[['primaryTitle', 'originalTitle', 'numVotes', 'averageRating']]
title_basics_ratings.head()
```

```python
# Display the 10 movies with the highest average rating
title_basics_ratings.sort_values('averageRating', ascending=False).head(10)
```

Obviously not accurate without a threshold on amount of votes

```python
# Display the 10 movies with the highest average rating and at least 100000 votes
title_basics_ratings[title_basics_ratings['numVotes'] >= 100000].sort_values('averageRating', ascending=False).head(10)
```
